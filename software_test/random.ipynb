```python
import random

# Panjang DNA
length = 5000

# Generate random DNA (A, T, G, C)
dna_seq = ''.join(random.choices("ATGC", k=length))

# Simpan ke file FASTA
fasta_path = "/Users/user/Downloads/Nicotiana_LLM/software_test/random_seq_test/random_5kb.fasta"
with open(fasta_path, "w") as f:
    f.write(">random_dna_2kb\n")  # header fasta
    # potong per 80 bp biar rapi sesuai format FASTA
    for i in range(0, length, 80):
        f.write(dna_seq[i:i+80] + "\n")

print(f"FASTA file saved to: {fasta_path}")
```

```python
import pandas as pd

# Path file CSV
file_path = "/Users/user/Downloads/Nicotiana_LLM/software_test/random_seq_test/promoter_igem.csv"

# Baca CSV
df = pd.read_csv(file_path)

# Tambah kolom 'length' berdasarkan panjang Sequence
df["length"] = df["Sequence"].astype(str).apply(len)

# Simpan hasil ke file baru (biar file asli tetap ada)
output_path = "/Users/user/Downloads/Nicotiana_LLM/software_test/random_seq_test/promoter_igem_with_length.csv"
df.to_csv(output_path, index=False)

print(f"Hasil disimpan di: {output_path}")
```

```python
import pandas as pd
import os

# Path file CSV input
csv_path = "/Users/user/Downloads/Nicotiana_LLM/software_test/random_seq_test/promoter_igem_with_length.csv"

# Folder output
output_dir = "/Users/user/Downloads/Nicotiana_LLM/software_test/random_seq_test"

# Baca CSV
df = pd.read_csv(csv_path)

# Buat 1 file FASTA per baris
for _, row in df.iterrows():
    name = str(row["Name"]).strip()
    length = row["length"]
    seq = str(row["Sequence"]).strip()

    fasta_path = os.path.join(output_dir, f"{name}.fasta")
    with open(fasta_path, "w") as f:
        f.write(f">{name}|length={length}\n{seq}\n")

print(f"Semua FASTA berhasil disimpan ke folder: {output_dir}")
```
